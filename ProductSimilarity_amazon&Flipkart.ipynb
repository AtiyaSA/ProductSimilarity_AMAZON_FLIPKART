```python
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from PIL import Image
import requests
from io import BytesIO
import time
from imagehash import average_hash
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings("ignore")
```

```python
inp = pd.read_excel("Scrappingproject.xlsx")
inp
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
```

```python
lt = []
# for i in range(inp.shape[0]):
for i in range(8,9):
    dic = {}
    dic['title'] = inp['Amazon_Product_Title'][i]
    dic['search_url'] = inp['Amazon_URL'][i]
    
#     dic['price'] = driver.find_element(By.ID, 'corePrice_desktop').text.split('\n')[1]
    link = []
    count = 3
    driver.get(dic['search_url'])
    driver.find_element(By.ID, 'imgTagWrapperId').click()
    img_l = []
    for k in range(7):
        try:
            driver.find_element(By.ID, 'ivImage_{}'.format(k)).click()
            time.sleep(2)
        except:
            break
        img_l.append(driver.find_element(By.ID, "ivLargeImage").find_element(By.TAG_NAME,'img').get_attribute('src'))
    for img_url in img_l:
        response = requests.get(img_url)
        img1 = Image.open(BytesIO(response.content))
        driver.get('https://www.google.com/imghp')
        try:
            driver.find_element(By.CLASS_NAME, 'QS5gu.sy4vM').click()
            time.sleep(2)
        except:
            pass
        driver.find_element(By.CLASS_NAME, 'nDcEnd').click()
        time.sleep(2)
        driver.find_element(By.CLASS_NAME, 'cB9M7').send_keys(img_url)
        driver.find_element(By.CLASS_NAME, 'Qwbd3').send_keys(Keys.ENTER)
        for l in driver.find_elements(By.CLASS_NAME,"Vd9M6.abDdKd.xuQ19b"):
            url = l.find_element(By.TAG_NAME, "a").get_attribute("href")
            if url.find("flipkart") != -1:
                link.append(url)
        if link!= []:
            break
    similar = 0.4
    for lnk in link:
        driver.get(lnk)
        img = driver.find_element(By.CLASS_NAME, "_2r_T1I._396QI4").get_attribute('src')
        response = requests.get(img)
        img2 = Image.open(BytesIO(response.content))
        hash1 = imagehash.average_hash(img1)
        hash2 = imagehash.average_hash(img2)
        # Compare the hash values
        dic['similarity'] = abs(1 - (hash1 - hash2) / len(hash1.hash))
        if similarity > similar:
            dic['flip_url'] = lnk
            similar = dic['similarity']
    driver.get(dic['flip_url'])
    dic['flip_title'] = driver.find_element(By.CLASS_NAME, "yhB1nd").text
    lt.append(dic)
```
